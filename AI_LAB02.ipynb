```python
# Initial list of cars
cars = ["BMW", "Audi", "Toyota", "Ford", "Honda"]
print("Initial cars list:", cars)
```

```python
cars.append("Mercedes")
print("After append():", cars)
```

```python
temp_cars = cars.copy()   # copy used to keep original list safe
temp_cars.clear()
print("After clear():", temp_cars)
```

```python
cars_copy = cars.copy()
print("Copied list:", cars_copy)
```

```python
cars.append("Audi")
print("Cars list:", cars)
print("Count of 'Audi':", cars.count("Audi"))
```

```python
more_cars = ["Hyundai", "Kia"]
cars.extend(more_cars)
print("After extend():", cars)
```

```python
cars.insert(2, "Tesla")
print("After insert():", cars)
```

```python
removed_car = cars.pop(3)
print("Popped car:", removed_car)
print("After pop():", cars)
```

```python
cars.remove("Ford")
print("After remove('Ford'):", cars)
```

```python
cars.reverse()
print("After reverse():", cars)
```

```python
cars.sort()
print("After sort():", cars)
```

```python
# GPA Calculator using Lists, Loop and If-Else

subjects = ["OOP", "DB", "Pak St.", "English"]
credits = [4, 3, 2, 2]

total_points = 0
total_credits = 0

for i in range(len(subjects)):
    marks = int(input("Enter marks for " + subjects[i] + ": "))

    if marks >= 85:
        gp = 4.0
    elif marks >= 75:
        gp = 3.5
    elif marks >= 65:
        gp = 3.0
    elif marks >= 55:
        gp = 2.5
    else:
        gp = 2.0

    total_points += gp * credits[i]
    total_credits += credits[i]

gpa = total_points / total_credits
print("Your GPA is:", round(gpa, 2))
```
