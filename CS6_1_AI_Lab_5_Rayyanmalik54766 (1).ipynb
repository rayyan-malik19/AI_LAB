#Dept First Search


```python
def DFS(node, graph, visited):
    visited[node] = True
    print("Visited:", node)
    print("Adjacent nodes of", node, ":", graph[node])

    for child in graph[node]:
        if not visited[child]:
            DFS(child, graph, visited)

if __name__ == "__main__":

    graph = {
         0: [1,2,3],
         1: [],
         2: [0,3,4],
         3: [0,2],
         4: [2]
    }

    visited = [False]*len(graph)

    for node in graph:
        if not visited[node]:
            DFS(node, graph, visited)
```
