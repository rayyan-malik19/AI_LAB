```python
def is_prime(n):
  if n==0 or n==1:
   return 0

  for i in range(2,n):
    if n % i ==0:
       return 0

       return 1
```

```python
x = int(input("enter a number"))
if is_prime(x)==0:
   print("not prime")
else:
  print("prime")
```
